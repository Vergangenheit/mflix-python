```python
lookup_stage = {
    "$lookup": {
          'from': 'comments',
          'localField': 'name',
          'foreignField': 'name',
          'as': 'commenter'
            }
}
add_field_stage = {
    "$addFields": {
      "number_of_comments": {"$size": "$commenter"}
        }
}
sort_stage = {
    "$sort": {
      "number_of_comments": -1
        }
}
unset_stage =  {
    "$unset": ['password', 'commenter']
}
pipeline = [lookup_stage, add_field_stage, sort_stage, unset_stage]
users_comments: CommandCursor = users.aggregate(pipeline)
list(users_comments)
```

```python
def add_comment(movie_id: str, user, comment: str, date: datetime.datetime) -> InsertOneResult:
    """
    Inserts a comment into the comments collection, with the following fields:

    - "name"
    - "email"
    - "movie_id"
    - "text"
    - "date"

    Name and email must be retrieved from the "user" object.
    """
    # TODO: Create/Update Comments
    # Construct the comment document to be inserted into MongoDB.
    comment_doc = {"name": user.name, "email": user.email, "movie_id": ObjectId(movie_id), "text": comment, "date": date}
    return comments.insert_one(comment_doc)
```

```python
def update_comment(comment_id: Union[ObjectId, str, bytes], user_email: str, text: str, date: datetime.datetime) -> Optional[UpdateResult]:
    """
    Updates the comment in the comment collection. Queries for the comment
    based by both comment _id field as well as the email field to doubly ensure
    the user has permission to edit this comment.
    """
    # TODO: Create/Update Comments
    # Use the user_email and comment_id to select the proper comment, then
    try:
        com: Dict = comments.find_one({"_id": comment_id})
        # update the "text" and "date" of the selected comment.
        response: UpdateResult = comments.update_one(
            {"_id": com.get("_id"), "email": user_email},
            {"$set": {"text": text, "date": date}}
        )
        return response
    except Exception as e:
        raise TypeError
```
